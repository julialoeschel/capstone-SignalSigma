```python
import pandas as pd
import numpy as np
from darts import TimeSeries
from darts.models import TFTModel
from darts.utils.likelihood_models import QuantileRegression
from darts.dataprocessing.transformers import Scaler
from darts.metrics import mae, rmse, smape, mape, r2_score, mase, mse, rmsle, marre
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import sys
import copy
from pytorch_lightning.callbacks.early_stopping import EarlyStopping

warnings.filterwarnings("ignore")

# === Step 1: Load data
print("[INFO] Loading data...")
df = pd.read_csv("../data/Stock_market_data/clean_stock_data_with_time_index.csv", parse_dates=["date"], index_col="date")
df = df.sort_index()
raw_main_df = copy.deepcopy(df)
print(f"[DEBUG] df shape: {df.shape}, index range: {df.index.min()} - {df.index.max()}, columns: {df.columns.tolist()}")

# === Step 2: Select target
print("[INFO] Selecting target column...")
target_stock = "AAPL"
drop_cols = [f"open_{target_stock}", f"low_{target_stock}", f"high_{target_stock}"]
df.drop(columns=drop_cols, inplace=True)

target_df = df[[f"close_{target_stock}"]].rename(columns={f"close_{target_stock}": "target"})
df.drop(columns=f"close_{target_stock}", inplace=True)
df_col = df.columns.tolist()
print(f"[DEBUG] target_df shape: {target_df.shape}, index: {target_df.index.min()} to {target_df.index.max()}")

# === Step 3: Time-based features
print("[INFO] Engineering time features...")
features = pd.DataFrame(index=target_df.index)
features["year"] = features.index.year
features["month"] = features.index.month
features["day_of_week"] = features.index.dayofweek
features["day_of_month"] = features.index.day
features["week"] = features.index.isocalendar().week.values
features["quarter"] = features.index.quarter

def encode_cyclical(df, col, max_val):
    df[f"{col}_sin"] = np.sin(2 * np.pi * df[col] / max_val)
    df[f"{col}_cos"] = np.cos(2 * np.pi * df[col] / max_val)
    df.drop(columns=[col], inplace=True)

for col, max_v in zip(["year", "month", "day_of_week", "day_of_month", "week", "quarter"],
                      [features.year.max(), 12, 7, 31, 53, 4]):
    encode_cyclical(features, col, max_v)

features = pd.concat([features, df[df_col]], axis=1)
print(f"[DEBUG] features shape: {features.shape}, columns: {features.columns.tolist()}")

# === Step 4: Merge & interpolate
print("[INFO] Merging and interpolating...")
full_df = pd.concat([target_df, features], axis=1).resample("D").ffill()
print(f"[DEBUG] full_df shape: {full_df.shape}, index range: {full_df.index.min()} - {full_df.index.max()}")

# === Step 5: TimeSeries conversion
print("[INFO] Converting to TimeSeries format...")
raw_target = TimeSeries.from_dataframe(full_df, value_cols="target", freq="D")
raw_covariates = TimeSeries.from_dataframe(full_df, value_cols=features.columns.tolist())
print(f"[DEBUG] raw_target len: {len(raw_target)}, type: {type(raw_target)}, start: {raw_target.start_time()}, end: {raw_target.end_time()}, freq: {raw_target.freq_str}, NaNs: {np.isnan(raw_target.values()).sum()}")
print(f"[DEBUG] raw_covariates len: {len(raw_covariates)}, shape: {raw_covariates.values().shape}, NaNs: {np.isnan(raw_covariates.values()).sum()}")

# === Step 6: Split
print("[INFO] Splitting data...")
output_len = 15
input_len = output_len * 6
val_size = min(((input_len + output_len)*2), int(0.1 * len(raw_target)))

train_target_raw = raw_target[:-val_size - output_len]
val_target_raw = raw_target[-val_size - output_len:-output_len]
test_target_raw = raw_target[-output_len:]

val_covariates_raw = raw_covariates.slice(val_target_raw.start_time(), val_target_raw.end_time() + pd.Timedelta(days=output_len))
test_covariates_raw = raw_covariates.slice(test_target_raw.start_time() - pd.Timedelta(days=input_len), test_target_raw.end_time() + pd.Timedelta(days=output_len))
train_covariates_raw = raw_covariates.slice(train_target_raw.start_time(), train_target_raw.end_time() + pd.Timedelta(days=output_len))

print(f"[DEBUG] Train target raw: len={len(train_target_raw)}, index=({train_target_raw.start_time()} to {train_target_raw.end_time()})")
print(f"[DEBUG] Val target raw: len={len(val_target_raw)}, index=({val_target_raw.start_time()} to {val_target_raw.end_time()})")
print(f"[DEBUG] Test target raw: len={len(test_target_raw)}, index=({test_target_raw.start_time()} to {test_target_raw.end_time()})")
print(f"[DEBUG] Train covariates raw: len={len(train_covariates_raw)}")
print(f"[DEBUG] Val covariates raw: len={len(val_covariates_raw)}")
print(f"[DEBUG] Test covariates raw: len={len(test_covariates_raw)}")

# === Step 7: Scaling
print("[INFO] Scaling data...")
t_scaler = Scaler()
f_scaler = Scaler()

train_target = t_scaler.fit_transform(train_target_raw)
val_target = t_scaler.transform(val_target_raw)
test_target = t_scaler.transform(test_target_raw)

train_covariates = f_scaler.fit_transform(train_covariates_raw)
val_covariates = f_scaler.transform(val_covariates_raw)
test_covariates = f_scaler.transform(test_covariates_raw)

print(f"[DEBUG] train_target shape: {train_target.values().shape}, NaNs: {np.isnan(train_target.values()).sum()} ")
print(f"[DEBUG] val_target shape: {val_target.values().shape}, NaNs: {np.isnan(val_target.values()).sum()} ")
print(f"[DEBUG] test_target shape: {test_target.values().shape}, NaNs: {np.isnan(test_target.values()).sum()} ")

print(f"[DEBUG] train_covariates shape: {train_covariates.values().shape}, NaNs: {np.isnan(train_covariates.values()).sum()} ")
print(f"[DEBUG] val_covariates shape: {val_covariates.values().shape}, NaNs: {np.isnan(val_covariates.values()).sum()} ")
print(f"[DEBUG] test_covariates shape: {test_covariates.values().shape}, NaNs: {np.isnan(test_covariates.values()).sum()} ")

# === Step 8: Define and Train TFT with EarlyStopping
print("[INFO] Initializing TFT model...")
model = TFTModel(
    input_chunk_length=input_len,
    output_chunk_length=output_len,
    hidden_size=64,
    lstm_layers=2,
    dropout=0.3,
    batch_size=64,
    n_epochs=2,
    num_attention_heads=4,
    force_reset=True,
    save_checkpoints=True,
    likelihood=QuantileRegression(quantiles=[0.1, 0.5, 0.9]),
    pl_trainer_kwargs={
        "callbacks": [EarlyStopping(monitor="val_loss", patience=3, mode="min")]
    }
)

try:
    print("[INFO] Training started...")
    model.fit(series=train_target, future_covariates=train_covariates,
              val_series=val_target, val_future_covariates=val_covariates,
              verbose=True)
except KeyboardInterrupt:
    print("[INFO] Training manually interrupted.")

# === Step 9: Predict
print("[INFO] Making predictions...")
train_pred = model.predict(n=output_len, series=train_target, future_covariates=train_covariates)

val_pred = model.predict(n=output_len, series=val_target, future_covariates=val_covariates)
test_pred = model.predict(n=output_len, series=val_target, future_covariates=test_covariates)

print(f"[DEBUG] train_pred shape: {train_pred.values().shape}, start: {train_pred.start_time()}, end: {train_pred.end_time()}, NaNs: {np.isnan(train_pred.values()).sum()}")
print(f"[DEBUG] val_pred shape: {val_pred.values().shape}, start: {val_pred.start_time()}, end: {val_pred.end_time()}, NaNs: {np.isnan(val_pred.values()).sum()}")
print(f"[DEBUG] test_pred shape: {test_pred.values().shape}, start: {test_pred.start_time()}, end: {test_pred.end_time()}, NaNs: {np.isnan(test_pred.values()).sum()}")



train_pred_inv = t_scaler.inverse_transform(train_pred)
val_pred_inv = t_scaler.inverse_transform(val_pred)
test_pred_inv = t_scaler.inverse_transform(test_pred)


print(f"[DEBUG] train_pred_inv shape: {train_pred_inv.values().shape}, start: {train_pred_inv.start_time()}, end: {train_pred_inv.end_time()}, NaNs: {np.isnan(train_pred_inv.values()).sum()}")
print(f"[DEBUG] val_pred_inv shape: {val_pred_inv.values().shape}, start: {val_pred_inv.start_time()}, end: {val_pred_inv.end_time()}, NaNs: {np.isnan(val_pred_inv.values()).sum()}")
print(f"[DEBUG] test_pred_inv shape: {test_pred_inv.values().shape}, start: {test_pred_inv.start_time()}, end: {test_pred_inv.end_time()}, NaNs: {np.isnan(test_pred_inv.values()).sum()}")

train_true = raw_target.slice(train_pred.start_time(), train_pred.end_time())
val_true = raw_target.slice(val_pred.start_time(), val_pred.end_time())
test_true = test_target_raw

print(f"[DEBUG] train_true shape: {train_true.values().shape}, start: {train_true.start_time()}, end: {train_true.end_time()}, NaNs: {np.isnan(train_true.values()).sum()}")
print(f"[DEBUG] val_true shape: {val_true.values().shape}, start: {val_true.start_time()}, end: {val_true.end_time()}, NaNs: {np.isnan(val_true.values()).sum()}")
print(f"[DEBUG] test_true shape: {test_true.values().shape}, start: {test_true.start_time()}, end: {test_true.end_time()}, NaNs: {np.isnan(test_true.values()).sum()}")



# === Step 10: Evaluation
print("[INFO] Evaluating model...")
def evaluate(true_ts, pred_ts, insample):
    print(f"[DEBUG] true_ts: shape={true_ts.values().shape}, type={type(true_ts)}")
    print(f"[DEBUG] pred_ts: shape={pred_ts.values().shape}, type={type(pred_ts)}")
    print(f"[DEBUG] insample: shape={insample.values().shape}, type={type(insample)}")
    return {
        "MAPE": mape(true_ts, pred_ts),
        "SMAPE": smape(true_ts, pred_ts),
        "MAE": mae(true_ts, pred_ts),
        "RMSE": rmse(true_ts, pred_ts),
        "R2": r2_score(true_ts, pred_ts),
        "RMSLE": rmsle(true_ts, pred_ts),
        "MSE": mse(true_ts, pred_ts)
    }

test_insample = raw_target.slice(train_target_raw.start_time(), test_target_raw.start_time() - pd.Timedelta(days=1))


test_metrics = evaluate(test_target_raw, test_pred_inv, test_insample)
performance_df = pd.DataFrame([test_metrics], index=["Test"])

from tabulate import tabulate
print(tabulate(performance_df, headers='keys', tablefmt='fancy_grid'))

# === Step 11: Quantile Forecasting
print("[INFO] Running quantile forecast...")
forecast = model.predict(n=output_len, series=val_target, future_covariates=test_covariates, num_samples=100)
forecast_p10 = t_scaler.inverse_transform(forecast.quantile_timeseries(0.1))
forecast_p50 = t_scaler.inverse_transform(forecast.quantile_timeseries(0.5))
forecast_p90 = t_scaler.inverse_transform(forecast.quantile_timeseries(0.9))

forecast_df = pd.DataFrame({
    "p10": forecast_p10.values().squeeze(),
    "p50": forecast_p50.values().squeeze(),
    "p90": forecast_p90.values().squeeze()
}, index=forecast_p50.time_index)

# === Step 12: Plot
print("[INFO] Plotting forecast...")
plt.figure(figsize=(12, 6))
forecast_p50.plot(label="Median (p50)")
forecast_p10.plot(label="Lower (p10)", linestyle="--")
forecast_p90.plot(label="Upper (p90)", linestyle="--")
test_true.plot(label="True", linewidth=2)
plt.title("Quantile Forecasting with TFT")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
